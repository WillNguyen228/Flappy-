```python
#Broadcast operation
v1 = torch.arange(4, dtype=torch.float32)
v2 = torch.tensor([3., 2., 1., 0.])

m1 = v1.reshape(2, 2)
m2 = v2.reshape(2, 2)

print(f"Vector 1: {v1}")
print(f"Vector 2: {v2}")
print(f"\nMatrix 1:\n{m1}")
print(f"\nMatrix 2:\n{m2}")

dot = torch.dot(v1, v2)
print(f"\nv1 · v2 = {dot}")

v3 = torch.tensor([3., 2.])
mv = torch.mv(m1, v3)
print(f"\nMatrix-vector multiplication (m1 @ v3): {mv}")

mm = torch.mm(m1, m2)
print(f"\nMatrix-matrix multiplication (m1 @ m2):\n{mm}")
```

```python
#Forward and backward passing

# Input features: 2 values [1.0, 2.0]
x = torch.tensor([1.0, 2.0])

# Target value we want to predict
y = torch.tensor(2.0)

# Learnable weights: requires_grad=True tells PyTorch to track gradients
w = torch.tensor([2.0, 0.5], requires_grad=True)

# Learning rate: controls how big each update step is
eta = 0.1

# FORWARD PASS: compute prediction
# y_hat = w[0]*x[0] + w[1]*x[1] = 2.0*1.0 + 0.5*2.0 = 3.0
y_hat = torch.dot(w, x)

# Compute loss: mean squared error
# Measures how far our prediction is from the target
loss = 0.5 * (y_hat - y) ** 2

print(f"Initial w: {w.tolist()}")
print(f"Prediction: {y_hat.item():.4f}")
print(f"Target: {y.item():.4f}")
print(f"Loss: {loss.item():.4f}")

# BACKWARD PASS: compute gradients
# Automatically calculates dLoss/dw for each weight
loss.backward()

print(f"Gradient dL/dw: {w.grad.tolist()}")

# UPDATE WEIGHTS: gradient descent step
# Move weights in opposite direction of gradient to reduce loss
with torch.no_grad():  # Don't track this operation for gradient computation
    w -= eta * w.grad

# Clear gradients for next iteration (they accumulate by default)
w.grad.zero_()

print(f"Updated w: {w.tolist()}")
```

```python
# Training Loop
# for epoch:
#    forward
#    loss
#    backward
#    update

# y = W x + b
# x = features
# W = weights
# y = outputs

# House price data: (square_feet_in_1000s, bedrooms, price_in_100k)
# Example: (1.0, 2, 2.0) means 1000 sqft, 2 bedrooms, $200k
data = [
    (1.0, 2, 2.0),
    (1.2, 2, 2.3),
    (1.5, 3, 2.85),
    (1.8, 3, 3.4),
    (2.0, 4, 4.0),
    (2.2, 4, 4.5),
    (2.5, 5, 5.2),
    (0.9, 1, 1.7),
]
x1 = torch.tensor([d[0] for d in data])
x2 = torch.tensor([float(d[1]) for d in data])
y_data = torch.tensor([d[2] for d in data])

w = torch.tensor([2.0, 0.5], requires_grad=True)
eta = 0.1

for step in range(100):
    y_hat = w[0] * x1 + w[1] * x2
    loss = 0.5 * (y_hat - y_data) ** 2
    loss = loss.mean()
    
    loss.backward()
    
    with torch.no_grad():
        w -= eta * w.grad
    w.grad.zero_()
    
    if step % 20 == 0:
        print(f"Step {step:3d}: loss = {loss.item():.6f}")

print(f"\nFinal weights: {w.tolist()}")
```

```python
# Linear Regression
import torch.nn as nn

# House price data: (square_feet_in_1000s, bedrooms, price_in_100k)
# Example: (1.0, 2, 2.0) means 1000 sqft, 2 bedrooms, $200k
data = [
    (1.0, 2, 2.0),
    (1.2, 2, 2.3),
    (1.5, 3, 2.85),
    (1.8, 3, 3.4),
    (2.0, 4, 4.0),
    (2.2, 4, 4.5),
    (2.5, 5, 5.2),
    (0.9, 1, 1.7),
]
sqft = torch.tensor([d[0] for d in data])
bedrooms = torch.tensor([float(d[1]) for d in data])
price = torch.tensor([d[2] for d in data])
n = len(price)

col_ones = torch.ones(n)
X = torch.stack([sqft, bedrooms, col_ones], dim=1)
y = price.reshape(-1, 1)

print(f"Design matrix X shape: {X.shape}")
print(f"Target vector y shape: {y.shape}")

XtX = X.T @ X
Xty = X.T @ y
w = torch.linalg.solve(XtX, Xty)

print(f"\nOptimal weights:")
print(f"  Square feet coefficient: {w[0].item():.3f}")
print(f"  Bedrooms coefficient: {w[1].item():.3f}")
print(f"  Bias term: {w[2].item():.3f}")

y_hat = X @ w
mse_loss = torch.mean((y_hat - y) ** 2)
print(f"\nMean Squared Error: {mse_loss.item():.4f}")
```
